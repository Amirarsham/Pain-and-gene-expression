```python
n=code
def primeFactors(n):
    fact=[]
    while n % 2 == 0:
        fact.append(2)
        n = n / 2
    for i in range(3,int(math.sqrt(n))+1,2):
            while n % i== 0:
                fact.append(i)
                n = n / i
    if n > 2:
        fact.append(n)
    return(fact)
   
a=[11, 13, 17, 19, "Row A"]
b=[23, 29, 31, 37, "Row B"]
c=[41, 43, 47, 53,"Row C"]
d=[59, 61, 67, 71,"Row D"]
```
